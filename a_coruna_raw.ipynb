```python
import requests
from bs4 import BeautifulSoup
import urllib.parse

# URL of the main page
base_url = 'https://www.coruna.gal/transparencia/es/normativa-municipal/normativa/ordenanzas'
# base_url = "https://www.coruna.gal/transparencia/es/normativa-municipal/normativa/ordenanzas/ordenanzas?_charset_=UTF-8&categoriaTipo=Contenido&categoriaID=1453773632416&textoBuscar=&buscar=Buscar"

def extract_hrefs(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all 'a' tags
        links = soup.find_all('a')

        # Extract href attributes
        hrefs = [link.get('href') for link in links if link.get('href')]

        return hrefs
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return []

def main():
    hrefs = extract_hrefs(base_url)

    # Print out all extracted hrefs
    for href in hrefs:
        # Convert relative URLs to absolute URLs
        absolute_url = urllib.parse.urljoin(base_url, href)
        print(absolute_url)

if __name__ == '__main__':
    main()
```

```python
def extract_and_follow_links(url):
    while url:
        print(f"Extracting from: {url}")
        hrefs = extract_hrefs(url)

        # Process or store hrefs as needed
        for href in hrefs:
            absolute_url = urllib.parse.urljoin(url, href)
            print(absolute_url)

        # Find the URL for the next page
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        next_button = soup.find('a', text='Next')  # Adjust selector if needed

        if next_button and next_button.get('href'):
            url = urllib.parse.urljoin(url, next_button.get('href'))
        else:
            url = None

if __name__ == '__main__':
    extract_and_follow_links(base_url)
```

```python
import requests
import pdfplumber
from io import BytesIO

# Function to download and extract text from a PDF URL
def extract_text_from_pdf(url):
    # Download the PDF
    response = requests.get(url)
    response.raise_for_status()  # Check for request errors

    # Open the PDF from bytes
    with BytesIO(response.content) as pdf_file:
        with pdfplumber.open(pdf_file) as pdf:
            # Extract text from all pages
            text = ''
            for page in pdf.pages:
                text += page.extract_text()

    return text

# Example URL of the PDF
pdf_url = 'https://www.coruna.gal/descarga/1453833567784/20230317%20Criterio%20interpretativo%20art%205%20Ordenanza.pdf'

# Extract and print text
pdf_text = extract_text_from_pdf(pdf_url)
print(pdf_text)
```

```python
import requests
import pdfplumber
from io import BytesIO
import pandas as pd
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from datetime import datetime

# Base URL and number of pages
base_url = "https://www.coruna.gal/obtenermasresultados/es?itemID=1453612770277&site=Transparencia"
num_pages = 9  # Number of pages to scrape

# Mapping category names to their corresponding categoriaID values
categories = {
    "Administración y gestión": "1453773633326",
    "Fiscales": "1453773634816",
    "Medio ambiente": "1453773637896",
    "Sanidad y consumo": "1453773640360",
    "Seguridad ciudadana": "1453773640660",
    "Servicios sociales": "1453773641299",
    "Transportes y circulación": "1453773642032",
    "Urbanismo": "1453773642338"
}

def extract_info(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    # Extracting all links
    links = [a['href'] for a in soup.find_all('a', href=True) if not a['href'].startswith('javascript:')]
    # Filter out links that end in .htm or .pdf
    filtered_links = [link for link in links if not link.lower().endswith(('.htm', '.pdf'))]
    # Prepend base URL to relative links
    full_links = [urljoin('https://www.coruna.gal/', link) for link in filtered_links]
    return full_links

def get_page_title(url):
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        title = soup.title.string if soup.title else 'No Title'
        return title.strip()
    except Exception as e:
        print(f"Error fetching title for URL {url}: {e}")
        return 'Error'

def extract_pdf_links_and_titles(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Check if the request was successful
        soup = BeautifulSoup(response.text, 'html.parser')
        pdf_links_with_titles = []

        # Find all <a> tags
        for a in soup.find_all('a', href=True):
            href = a['href']
            if href.lower().endswith('.pdf'):
                pdf_url = urljoin(url, href)
                # Extract the descriptive text
                pdf_title = a.get_text(strip=True)
                pdf_links_with_titles.append((pdf_url, pdf_title))

        return pdf_links_with_titles
    except requests.RequestException as e:
        print(f"Request failed for {url}: {e}")
        return []

def extract_text_from_pdf(url):
    try:
        # Download the PDF
        response = requests.get(url)
        response.raise_for_status()  # Check for request errors

        # Open the PDF from bytes
        with BytesIO(response.content) as pdf_file:
            with pdfplumber.open(pdf_file) as pdf:
                # Extract text from all pages
                text = ''
                for page in pdf.pages:
                    text += page.extract_text() or ''  # Handle pages with no text

        return text
    except Exception as e:
        print(f"Error extracting text from PDF {url}: {e}")
        return 'Error'

# Initialize lists to hold category, page numbers, URLs, titles, PDF URLs, and PDF titles
category_list = []
page_numbers = []
all_urls = []
titles = []
pdf_urls = []
pdf_titles = []
pdf_texts = []

# Get current date
current_date = datetime.today().strftime('%Y-%m-%d')

# Iterate over all categories and pages
for category, categoria_id in categories.items():
    for page_num in range(1, num_pages + 1):
        page_url = f"{base_url}&categoriaID={categoria_id}&argPag={page_num}#Componente1453612770277"
        print(f"Scraping category '{category}' page {page_num}...")
        urls = extract_info(page_url)
        category_list.extend([category] * len(urls))
        page_numbers.extend([page_num] * len(urls))
        all_urls.extend(urls)
        # Get titles for the URLs
        for url in urls:
            titles.append(get_page_title(url))

# Create a DataFrame
df = pd.DataFrame({
    'Category': category_list,
    'Page Number': page_numbers,
    'URL': all_urls,
    'Title': titles,
    'ciudad': 'A Coruña',
    'date': current_date
})

# Initialize lists to hold the expanded rows
expanded_data = []

# Iterate over the DataFrame and extract PDF links and titles
for index, row in df.iterrows():
    url = row['URL']
    pdfs_with_titles = extract_pdf_links_and_titles(url)
    if not pdfs_with_titles:
        # If no PDFs are found, just append the row as is
        expanded_data.append({
            'Category': row['Category'],
            'Page Number': row['Page Number'],
            'URL': row['URL'],
            'Title': row['Title'],
            'PDF URL': None,
            'PDF Title': None,
            'PDF Text': None,
            'ciudad': 'A Coruña',
            'date': current_date
        })
    else:
        # If PDFs are found, create a new row for each PDF
        for pdf_url, pdf_title in pdfs_with_titles:
            pdf_text = extract_text_from_pdf(pdf_url)
            expanded_data.append({
                'Category': row['Category'],
                'Page Number': row['Page Number'],
                'URL': row['URL'],
                'Title': row['Title'],
                'PDF URL': pdf_url,
                'PDF Title': pdf_title,
                'PDF Text': pdf_text,
                'ciudad': 'A Coruña',
                'date': current_date
            })

# Create a DataFrame for the expanded data
expanded_df = pd.DataFrame(expanded_data)

# Optionally, save to a CSV file
expanded_df.to_csv('expanded_data_with_pdfs_and_text.csv', index=False)

print("Data scraping complete. Data saved to 'expanded_data_with_pdfs_and_text.csv'.")
```
